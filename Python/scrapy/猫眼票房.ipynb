```python
import requests
import csv
from pyecharts.charts import Bar
from pyecharts import options as ops

url = 'http://piaofang.maoyan.com/second-box?beginDate=20200110'

res = requests.get(url)

josn = res.json()

print(josn['data']['list'][0])

bar = Bar()
movieNameList = []
boxInfoList = []
with open('./csv/20200111.csv', 'w') as csvfile:
    fieldnames = ['id', 'movieName', '上座率','拍片占比','票房占比','综合票房（万）']
    writer = csv.DictWriter(csvfile,fieldnames=fieldnames)
    writer.writeheader()
    index = 0
    for i in josn['data']['list']:
#         print(i['movieId'])
        if index < 10:
            movieNameList.append(i['movieName'])
            boxInfoList.append(i['boxInfo'])
        index += 1
        row = {}
        row['id'] = i['movieId']
        row['movieName'] = i['movieName']
        row['上座率'] = i['avgSeatView']
        row['拍片占比'] = i['showRate']
        row['票房占比'] = i['boxRate']
        row['综合票房（万）'] = i['boxInfo']
        writer.writerow(row)

bar.add_xaxis(movieNameList)
bar.add_yaxis('综合票房（万）',boxInfoList)
bar.set_global_opts(title_opts=ops.TitleOpts(title = '19年圣诞节票房排行'))
bar.render_notebook()
```

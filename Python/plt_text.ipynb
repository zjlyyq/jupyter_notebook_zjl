### Axes
轴包含两个（如果是3D，则包含三个）Axis对象（请注意轴和Axis之间的区别）。
Axes具有以下功能：
每个轴都有一个标题（通过set_title（）设置），一个x标签（通过set_xlabel（）设置）和一个y标签（通过set_ylabel（）设置）

```python
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
print(axs)
fig, ax = plt.subplots()  # a figure with a single Axes
axs[0][0].plot([1, 2, 3, 5], [1, 2, 3, 4])
axs[0][1].plot([1, 2, 3, 5], [4, 2, 1, 4])
axs[1][0].plot([1, 2, 3, 5], [0, 2, 1, 4])
axs[1][1].plot([1, 2, 3, 5], [0, 2, 0, 4], color="#fd3200")
print(ax)
ax.set_title('the most common move')
ax.set_xlabel('move time')
ax.set_ylabel('position')
ax.plot([1, 2, 3, 5], [1, 4, 2, 4], color="green")
```

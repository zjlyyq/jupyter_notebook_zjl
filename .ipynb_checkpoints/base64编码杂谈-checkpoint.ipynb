字符串‘Hw'的二进制数据是01001000,01100101

```python
# 字符串'He'的ASCII码分别是72，101
num = [72,104]

# 字符串翻转
def stringRsverse(str):
    ans = ''
    index = len(str) - 1
    while(index >= 0):
        ans += str[index]
        index = index -1
    return ans
    
# 数字转为二进制字符串，72转换为01001000
def intToB(num):
    list = ''
    while(num):
        list = list + str(num%2)
        num = num // 2
    offset = 8-len(list)
#     print('offset = '+str(offset))
    while(offset>0):
        offset = offset - 1
        list = list + '0'
#     print(list)
#     print(stringRsverse(list))
    return stringRsverse(list)
# 一个数组转换成二进制字符串拼接起来，每个二进制字符串用','分割
def main(list):
    ans = ''
    for num in list:
        ans = ans + intToB(num)
        if num != list[len(list)-1]:
            ans = ans + ','
    print(ans)
    return ans
result = main([72,101])
# print('resule = '+result)
```
